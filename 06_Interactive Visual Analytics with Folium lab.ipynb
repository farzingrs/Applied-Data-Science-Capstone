```python
# TASK 3: Calculate the distances between a launch site to its proximities
def calculate_distance(lat1, lon1, lat2, lon2):
    R = 6373.0  # approximate radius of earth in km
    lat1 = radians(lat1)
    lon1 = radians(lon1)
    lat2 = radians(lat2)
    lon2 = radians(lon2)
    dlon = lon2 - lon1
    dlat = lat2 - lat1
    a = sin(dlat / 2)**2 + cos(lat1) * cos(lat2) * sin(dlon / 2)**2
    c = 2 * atan2(sqrt(a), sqrt(1 - a))
    distance = R * c
    return distance
```

```python
from math import sin, cos, sqrt, atan2, radians

def calculate_distance(lat1, lon1, lat2, lon2):
    # approximate radius of earth in km
    R = 6373.0

    lat1 = radians(lat1)
    lon1 = radians(lon1)
    lat2 = radians(lat2)
    lon2 = radians(lon2)

    dlon = lon2 - lon1
    dlat = lat2 - lat1

    a = sin(dlat / 2)**2 + cos(lat1) * cos(lat2) * sin(dlon / 2)**2
    c = 2 * atan2(sqrt(a), sqrt(1 - a))

    distance = R * c
    return distance
```

```python
# Loop through all launch sites to calculate distances to proximities
for index, site in launch_sites_df.iterrows():
    launch_site_lat = site['Lat']
    launch_site_lon = site['Long']
    launch_site_name = site['Launch Site']

    # Example proximity coordinates (adjust using MousePosition)
    coastline_lat, coastline_lon = launch_site_lat + 0.0005, launch_site_lon + 0.005  # Approximate coastline
    railway_lat, railway_lon = launch_site_lat + 0.009, launch_site_lon - 0.008     # Approximate railway
    highway_lat, highway_lon = launch_site_lat - 0.001, launch_site_lon + 0.007     # Approximate highway
    city_lat, city_lon = launch_site_lat + 0.05, launch_site_lon - 0.23            # Approximate city

    # Calculate distances
    distance_coastline = calculate_distance(launch_site_lat, launch_site_lon, coastline_lat, coastline_lon)
    distance_railway = calculate_distance(launch_site_lat, launch_site_lon, railway_lat, railway_lon)
    distance_highway = calculate_distance(launch_site_lat, launch_site_lon, highway_lat, highway_lon)
    distance_city = calculate_distance(launch_site_lat, launch_site_lon, city_lat, city_lon)

    print(f"\nDistances for {launch_site_name}:")
    print(f"Distance to coastline: {distance_coastline:.2f} KM")
    print(f"Distance to railway: {distance_railway:.2f} KM")
    print(f"Distance to highway: {distance_highway:.2f} KM")
    print(f"Distance to city: {distance_city:.2f} KM")

    # Add Markers for each proximity
    for coord, distance, label in [
        ([coastline_lat, coastline_lon], distance_coastline, 'Coastline'),
        ([railway_lat, railway_lon], distance_railway, 'Railway'),
        ([highway_lat, highway_lon], distance_highway, 'Highway'),
        ([city_lat, city_lon], distance_city, 'City')
    ]:
        marker = folium.Marker(
            coord,
            icon=DivIcon(
                icon_size=(20, 20),
                icon_anchor=(0, 0),
                html='<div style="font-size: 12; color:#d35400;"><b>%s: %s KM</b></div>' % (label, "{:10.2f}".format(distance)),
            )
        )
        site_map.add_child(marker)

        # Add PolyLine
        lines = folium.PolyLine(locations=[[launch_site_lat, launch_site_lon], coord], weight=1, color='blue')
        site_map.add_child(lines)

# Save the updated map to HTML
site_map.save('all_proximities_map.html')
print("\nMap saved as 'all_proximities_map.html'. Open this file in a browser to view the map.")

# Try to display the map
display(site_map)
```

```python
# Calculate average distance to cities for all launch sites
distances_to_city = []
for index, site in launch_sites_df.iterrows():
    launch_site_lat = site['Lat']
    launch_site_lon = site['Long']
    city_lat, city_lon = launch_site_lat + 0.05, launch_site_lon - 0.23  # Approximate city
    distance_city = calculate_distance(launch_site_lat, launch_site_lon, city_lat, city_lon)
    distances_to_city.append(distance_city)

average_distance_to_city = sum(distances_to_city) / len(distances_to_city)
print(f"Average distance to cities across all launch sites: {average_distance_to_city:.2f} KM")
```
